Features are computed from a digitized image of a fine needle
aspirate (FNA) of a breast mass.  They describe
characteristics of the cell nuclei present in the image.

Separating plane described above was obtained using
Multisurface Method-Tree (MSM-T) [K. P. Bennett, "Decision Tree
Construction Via Linear Programming." Proceedings of the 4th
Midwest Artificial Intelligence and Cognitive Science Society,
pp. 97-101, 1992], a classification method which uses linear
programming to construct a decision tree.  Relevant features
were selected using an exhaustive search in the space of 1-4
features and 1-3 separating planes.

The actual linear program used to obtain the separating plane
in the 3-dimensional space is that described in:
[K. P. Bennett and O. L. Mangasarian: "Robust Linear
Programming Discrimination of Two Linearly Inseparable Sets",
Optimization Methods and Software 1, 1992, 23-34].

## Preprocessing

```python
#Import the dataset again for the clean start of the preprocessing stage
dataset = pd.read_csv('data.csv')
```

```python
#Getting data from the diagnosis column (our targets)
targets = dataset.diagnosis
```

```python
#Counting each value in the targets variable
targets.value_counts()
```

### Normalizing targets
Change target from strings to numbers

```python
targets = np.where(targets.values == 'M', 0, 1)
```

```python
dataset.head() #We don't need [id, diagnosis, Unnamed: 32] columns
```

```python
list = ['Unnamed: 32','id','diagnosis']
data = dataset.drop(list,axis = 1 )
data.shape
```

### Normalizing features

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
data.head()
```

```python
data = scaler.fit_transform(data)
```

```python
data
```

### Splitting data into training and test parts

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, targets, test_size=0.1)
```

```python
X_test.shape
```

### KNN using sklearn library

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

## Grid search

```python
from sklearn.model_selection import GridSearchCV
```

Parameters for the KNN

```python
param_grid = {'n_neighbors':[5, 7, 11, 15, 3], 
              'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'], 
              'leaf_size':[10, 30, 50, 100]}
```

#### Define the GridSearch for the KNN

```python
grid_search = GridSearchCV(knn, param_grid=param_grid)
```

#### Train the Grid Search

```python
grid_search.fit(X_train, y_train)
```

### Training the best KNN model

#### Get best set of parameters from the Grid Search

```python
grid_search.best_estimator_
```

#### Define the KNN

```python
knn = KNeighborsClassifier(algorithm='auto', leaf_size=10, metric='minkowski',
           metric_params=None, n_jobs=1, n_neighbors=5, p=2,
           weights='uniform')
```

#### Train the KNN

```python
knn.fit(X_train, y_train)
```

## Evaluating the model

### Accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
predict = knn.predict(X_test)
```

```python
print("Our algorithm has accuracy of: {}".format(accuracy_score(y_test, predict)))
```

### Confusion Matrix

    TRUE POSITIVES (TP): Cases when we predict that a patient has Malignant cancer and patient HAS the Malignant cancer. 
    TRUE NEGATIVES (TN): Cases when we predict that a patient has Benign cancer and patient HAS the Benign cancer.
    FALSE POSITIVES (FP): Cases when we predict that a patient has Malignant cancer and patient DOES NOT HAVE  the Malignant cancer.  (Known as a "Type I error.")
    FALSE NEGATIVES (FN): Cases when we predict that a patient has Benign cancer and patient HAS the Malignant cancer. (Known as a "Type II error.")

```python
from sklearn.metrics import confusion_matrix
```

```python
matrix = confusion_matrix(y_test, predict)
```

```python
sns.heatmap(matrix,annot=True,fmt="d")
```

### Precision score

Precision measures how accurate a model is out of those predicted positive, how many of them are actual positive.

Precision is a good measure to determine, when the costs of False Positive is high. 

When a cancer is Benign and indentified as Malignant, it is not very bad situation because the patient will get checked, but still we must make sure to avoide these situations.

```python
from sklearn.metrics import precision_score
```

```python
precision = precision_score(y_test, predict)
```

```python
print("Precision for our model is: ", precision)
```

### Recall

Recall calculates how many of the Actual Positives our model capture through labeling it as Positive (True Positive).

Based on this understanding, the Recall shall be the model metric we use to select our best model when there is a high cost associated with False Negative (our case)!

When a cancer is Malignant and indentified as Benign, we must make sure not to have these situations!

```python
from sklearn.metrics import recall_score
```

```python
recall = recall_score(y_test, predict)
```

```python
print("Recall for our model is: ", recall)
```
